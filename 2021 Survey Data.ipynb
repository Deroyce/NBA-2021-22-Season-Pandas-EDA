### Find most popular coding language for black coders

```python
all_black_df['LanguageHaveWorkedWith'].value_counts(dropna=False)

# 7 null values for languageHaveWorkedWith
```

```python
# all rows except for the NA's under LanguageHaveWorkedWith
language_df = all_black_df.loc[~all_black_df['LanguageHaveWorkedWith'].isnull()]
```

##### Have to split up the languages worked with entry for each person surveyed in order to count the individual languages.  Use a counter to collect this info

```python
from collections import Counter
language_counter = Counter()
```

```python
#make dataframe for just LanguageHaveWorkedWith column
NEWlanguage_df = language_df['LanguageHaveWorkedWith']
NEWlanguage_df.head()
```

```python
for response in NEWlanguage_df:
    language_counter.update(response.split(';'))
```

```python
language_counter.most_common(20)
```

```python
languages=[]
popularity=[]

for item in language_counter.most_common(20):
    languages.append(item[0])
    popularity.append(item[1])
```

```python
#Find out how to make a dataframe out of the languages list and popularity list

dflanguage = {'language':languages,'Count':popularity}
dflanguage2=pd.DataFrame(dflanguage)
dflanguage2
```

```python
#plot the newly created language dataframe from above
plt.figure(figsize=(20,6))
language_bar =sns.barplot(data=dflanguage2,x='language',y='Count')
language_bar.set_xticklabels(language_bar.get_xticklabels(),rotation = 80)
language_bar.set_title('Black Coders Most Popular Language')
language_bar
```

According to the survey, JavaSript is the most popular languge chosen by black coders with R being the least popular
