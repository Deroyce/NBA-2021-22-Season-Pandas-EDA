### Visualize the data

```python
def boxtotal(stat): #box plot for various metic totals
    med_order=dftotals4.groupby(by=['Pos'])[stat].median().sort_values(ascending=False).index  #orders plots by median
    
    plt.figure(figsize=(35,20))
    nbabox = sns.boxplot(data=dftotals4, x='Pos', y=stat, order=med_order)
    sns.set(font_scale=3)
    nbastrip = sns.stripplot(data=dftotals4, x='Pos', y=stat, color='.7', order=med_order)
    nbabox.set_title('Positions ('+ stat +') - Total')
    nbabox
    nbastrip

def boxaverage(stat):  #box plot for various metic averages
    med_order=dftotals4.groupby(by=['Pos'])[stat].median().sort_values(ascending=False).index  #orders plots by median

    plt.figure(figsize=(35,20))
    nbabox = sns.boxplot(data=df4, x='Pos', y=stat, order=med_order)
    sns.set(font_scale=3)
    nbastrip = sns.stripplot(data=df4, x='Pos', y=stat, color='.7', order=med_order)
    nbabox.set_title('Positions ('+ stat +') - Average')
    nbabox
    nbastrip
```

### Heat map of team totals

```python
dftotals4.columns
```

```python
import numpy as np

#create pivot table
team_pivot = pd.pivot_table(dftotals4,index=['Tm'], values=['Age','3P','3PA','2P','2PA','FT', 'FTA','ORB', 'DRB', 'TRB','AST','PTS','BLK','STL','TOV'], 
               aggfunc={'Age':np.mean,'3P':np.sum,'3PA':np.sum,'2P':np.sum,'2PA':np.sum,'FT':np.sum,'FTA':np.sum,
                        'ORB':np.sum,'DRB':np.sum,'TRB':np.sum,'AST':np.sum,'PTS':np.sum,'BLK':np.sum,'STL':np.sum,'TOV':np.sum})
```

```python
team_pivot
```

 Have to normalize the pivot table data across metrics in order for the heat map to show the correct scales

```python
teamdf = team_pivot.reset_index() #make the pivot table into a dataframe
teamdf_normalized = teamdf.copy()
```

```python
teamdf_normalized.drop(['Tm'], axis=1, inplace=True) #drop the 'Tm' column before normalizing
```

```python
for cols in teamdf_normalized.columns:
    teamdf_normalized[cols] = (teamdf_normalized[cols]-teamdf_normalized[cols].min())/(teamdf_normalized[cols].max()-teamdf_normalized[cols].min())
```

```python
teamdf_normalized2 = pd.concat([teamdf['Tm'],teamdf_normalized], axis=1) #combine the team names back with the normalized data
```

```python
teamdf_normalized2.set_index('Tm',inplace=True) #make 'Tm' the index for chart purposes
```

```python
plt.figure(figsize=(45,30))
sns.set(font_scale=2.5)
team_heatmap = sns.heatmap(teamdf_normalized2, cmap='BuPu',linecolor='white',linewidth=2,annot=True)
team_heatmap.set_title('NBA Team Stats')
```

Above figures in heat map have been normalized.  For each stat, the numbers range from 0 (least) to 1(most). Below are some observations.
